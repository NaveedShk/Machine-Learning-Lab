```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("cancer patient data sets.csv")
```

```python
print("Unique categories in Target Variable:", df['Patient Id'].unique())
```

```python
print("Total Columns:", len(df.columns))
print("\nFirst 50 Records:\n", df.head(50))
print("\nLast 30 Records:\n", df.tail(30))

print("\nDataset Info:")
print(df.info())

print("\nDataset Shape:", df.shape)  
print("Dataset Dimensions:", df.ndim)
print("Dataset Memory Usage:\n", df.memory_usage(deep=True))
```

```python
num_rows, num_cols = df.shape
print("Number of Samples (Rows):", num_rows)
print("Number of Features (Columns):", num_cols)
```

```python
print(df['Patient Id'].value_counts())

# Bar plot
sns.countplot(x='Patient Id', data=df)
plt.title("Distribution of Target Variable")
plt.show()
```

```python
subset_loc = df.loc[550:900, "Gender":"Coughing of Blood"]


subset_loc.to_csv("subset_loc.csv", index=False)

print("Mean:\n", subset_loc.mean(numeric_only=True))
print("\nMedian:\n", subset_loc.median(numeric_only=True))
print("\nMode:\n", subset_loc.mode().iloc[0])
print("\nStandard Deviation:\n", subset_loc.std(numeric_only=True))
print("\nMin Values:\n", subset_loc.min(numeric_only=True))
print("\nMax Values:\n", subset_loc.max(numeric_only=True))
```

```python
subset_iloc = df.iloc[-20:, df.columns.get_loc("Obesity"):df.columns.get_loc("Weight Loss")+1]


subset_iloc.to_csv("subset_iloc.csv", index=False)


print("Mean:\n", subset_iloc.mean(numeric_only=True))
print("\nMedian:\n", subset_iloc.median(numeric_only=True))
print("\nMode:\n", subset_iloc.mode().iloc[0])
print("\nStandard Deviation:\n", subset_iloc.std(numeric_only=True))
print("\nMin Values:\n", subset_iloc.min(numeric_only=True))
print("\nMax Values:\n", subset_iloc.max(numeric_only=True))
```

```python
specific_rows = df.loc[[110,220,360,440,656,778,202,889], :]


avg_alcohol_use = specific_rows['Alcohol use'].mean()
print("Average Alcohol Use in selected records:", avg_alcohol_use)
```

```python
sns.countplot(x='Gender', data=df)
plt.title("Distribution of Gender")
plt.show()


df['Age'].hist(bins=20, figsize=(6,4))
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()


sns.countplot(x='Level', data=df, palette="viridis")
plt.title("Level Distribution")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Heatmap")
plt.show()
```

```python
plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
sns.histplot(df['Age'], kde=True, bins=20, color="blue")
plt.title("Distribution of Age")

plt.subplot(1,2,2)
sns.countplot(x='Smoking', data=df, palette="Set2")
plt.title("Distribution of Smoking")

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
sns.boxplot(x=df['Obesity'], color="orange")
plt.title("Outliers in Obesity (Boxplot)")


plt.subplot(1,2,2)
sns.histplot(df['Obesity'], kde=True, bins=15, color="green")
plt.title("Obesity Distribution")

plt.tight_layout()
plt.show()
```
